```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
t, timestep = np.linspace(0,10,10000, retstep=True)
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(sp.size, d=timestep)

```

```python
freq.size
```

```python
fig, ax = plt.subplots()
ax.plot(t, np.sin(t))

fig, ax = plt.subplots()
ax.plot(freq, sp.real)
ax.plot(freq, sp.imag)
ax.set_xlim(-5, 5)
```

```python
from scipy.fftpack import fft, ifft, fftfreq, fftshift
```

```python
time, timestep = np.linspace(1, 5, 5, retstep = True)
x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
y = fft(x)
f = fftfreq(time.size, timestep)
f = fftshift(f)

yinv = ifft(y)
```

```python
fig, ax = plt.subplots()
ax.plot( np.arange(x.size), x)
ax.plot( np.arange(x.size), yinv)

fig, ax = plt.subplots()
ax.plot(f, y)
```
