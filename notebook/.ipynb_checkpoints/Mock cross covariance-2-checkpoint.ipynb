```python
def invert_covtot( covP, covXi, covPXi ):
    
    b = covPXi
    c = covPXi.T #matrix[cutInd+1:, 0:cutInd+1]
    d = covXi
    #ia = FisherP #masking(RSDPower, FisherP, kmin=kmin, kmax=kmax)
    ia = np.zeros(covP.shape)
    np.fill_diagonal(ia, 1./covP.diagonal())
    #ia = np.linalg.inv(covP)
    
    Fd = pinv( d - np.dot( np.dot( c, ia ), b) )
    Fc = - np.dot( np.dot( Fd, c), ia)
    Fb = - np.dot( np.dot( ia, b ), Fd )
    Fa = ia + np.dot( np.dot (np.dot( np.dot( ia, b), Fd ), c), ia)

    Fisher3_tot = np.vstack(( np.hstack(( Fa, Fb )), np.hstack(( Fc, Fd )) ))
    Fisher3_tot = (Fisher3_tot + Fisher3_tot.T) * 0.5
    print 'Fisher3_tot diagonal test', np.sum(Fisher3_tot.diagonal() <0)
```

```python
import scipy as sp
import scipy.interpolate

def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = sp.interpolate.interp1d(logx, logy, kind=kind)
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp
```

```python
def mock_covariance(p_model, xi_model, p_mock, xi_mock):
    m1, m2 = np.mgrid[0:p_model.size, 0:xi_model.size]
    mock_covpxi = np.zeros(( p_model.size, xi_model.size ))
    for p, x in zip(p_mock, xi_mock) :
        mock_covpxi += ( p[m1] - p_model[m1]) * (x[m2] - xi_model[m2])
    mock_covpxi = 1./( len(p_mock) - 1 ) * mock_covpxi
    
    if p_model.size == xi_model.size : mock_covpxi = (mock_covpxi + mock_covpxi.T)/2.
    return mock_covpxi
```

```python
def generate_mocks( RSDPower, p_model, covP_model, size = 500 ):
    print 'generate mocks... size=', size
    p_mock = np.random.multivariate_normal(p_model, covP_model, size=size)
    xi_mock = np.array([ RSDPower.fourier_transform_kr(0, p) for p in p_mock])
    
    return p_mock, xi_mock
    
def save_mocks( RSDPower, p_model, covP_model, size = 500, header = '', dir = '../data_txt/mocks/'  ):
    
    if not os.path.exists(dir) : os.makedirs(dir)
    p_mock, xi_mock = generate_mocks( RSDPower, p_model, covP_model, size = size )
    
    DAT_p = np.column_stack(( p_mock ))
    DAT_xi = np.column_stack((xi_mock))
    #import os
    #os.mkdir('../data_txt/mocks/')
    np.savetxt(dir + 'mocks_p.dat', DAT_p, header = header)
    np.savetxt(dir + 'mocks_xi.dat', DAT_xi,header = header)
    np.savetxt(dir + 'r.dat', RSDPower.rcenter,header = header)
    np.savetxt(dir + 'k.dat', RSDPower.kcenter_y,header = header)
    
def load_mocks( dir = '../data_txt/mocks/'  ):
    mocks_p = np.loadtxt(dir + 'mocks_p.dat')
    mocks_xi = np.loadtxt(dir + 'mocks_xi.dat')
    rcenter = np.loadtxt(dir + 'r.dat')
    kcenter = np.loadtxt(dir + 'k.dat')
    return kcenter, mocks_p, rcenter, mocks_xi
```

```python
from mcmc_lib import *
def compute_data_vector(lik_class, b=2.0,f=0.74,s=3.5):
    lik_class.b = b
    lik_class.f = f
    lik_class.s = s
    #datav = P_multipole(lik_class)
    datavp = lik_class.multipole_P(0)
    datavxi = lik_class.multipole_Xi(0)
    return [datavp, datavxi]

def best_chisqr( covinv = None, datav=None, mockv = None, p=False):
    d_diff = (datav - mockv)
    if p : chisqr = np.sum(d_diff**2 * covinv.diagonal())
    else : chisqr = np.dot(np.dot( d_diff, covinv ), d_diff)
    return chisqr
```

```python
def getting_sigma_bs( RSDPower, cov = None, datavs = None, 
                     mockdatavs = None, p=False, kmin=None, kmax = None,
                     rmin = None, rmax = None):
    
    blist = np.linspace(1.5, 2.5, 100)
    
    if kmin == None : 
        idx_kmin = 0
        idx_kmax = RSDPower.kmax_y.size
    else : 
        idx_kmin = get_closest_index_in_data( kmin, RSDPower.kmin_y )   
        idx_kmax = get_closest_index_in_data( kmax, RSDPower.kmax_y )
    
    if rmin == None : 
        idx_rmin = RSDPower.rmax.size
        idx_rmax = 0
    else : 
        idx_rmin = get_closest_index_in_data( rmin, RSDPower.rmin )   
        idx_rmax = get_closest_index_in_data( rmax, RSDPower.rmax )
        
    mask = np.zeros(cov.shape[0], dtype=bool)
    if p : mask[idx_kmin:idx_kmax+1] = 1
    else : mask[idx_rmax:idx_rmin+1] = 1
        
    Nk = np.sum(mask)
    m1, m2 = np.mgrid[0:cov.shape[0],0:cov.shape[0]]
    mask_2d = mask[m1] * mask[m2]
           
    if p :print 'Nk ', Nk, ' kmin', idx_kmin, ' kmax', idx_kmax
    else : print 'Nr ', Nk, ' rmin', idx_rmin, ' rmax', idx_rmax
    
    covinv = np.linalg.inv(cov[mask_2d].reshape(Nk,Nk))
    fig, ax = plt.subplots()
    
    bestfit_b = []
    chi2result = np.zeros(blist.size)
    for i in range(mockdatavs.shape[1]):
    #for i in range(1):
        print '{}/{} \r'.format(i+1, mockdatavs.shape[1] ),
        for j in range(blist.size) : 
            dv = datavs[j][mask]
            mv = mockdatavs[:,i][mask]           
            chi2result[j] = best_chisqr( covinv = covinv, datav=dv, mockv = mv, p=p)
            #print dv.shape,  mv.shape, chi2result[j]
        maxind = np.argmin(chi2result)
        #print maxind
        bestfit_b.append(blist[maxind])
        ax.plot(blist, chi2result, color = 'grey', alpha = 0.2)
        
        
    print ' sigma_b :', np.std(bestfit_b)
    
    
    return bestfit_b



def getting_sigma_bs_com( RSDPower, cov = None, datavs = None, mockdatavs = None, p=False, 
                         kmin=None, kmax = None, rmin=None, rmax = None ):
    
    blist = np.linspace(1.5, 2.5, 100)
    
    if kmin == None : 
        idx_kmin = 0
        idx_kmax = RSDPower.kmax_y.size
    else : 
        idx_kmin = get_closest_index_in_data( kmin, RSDPower.kmin_y )   
        idx_kmax = get_closest_index_in_data( kmax, RSDPower.kmax_y )
        
    if rmin == None : 
        idx_rmin = RSDPower.rmax.size
        idx_rmax = 0
    else : 
        idx_rmin = get_closest_index_in_data( rmin, RSDPower.rmin )   
        idx_rmax = get_closest_index_in_data( rmax, RSDPower.rmax )
        
        
    maskp = np.zeros(RSDPower.kcenter_y.size, dtype=bool)
    maskp[idx_kmin:idx_kmax+1] = 1
    maskxi = np.zeros(RSDPower.rcenter.size, dtype=bool)
    maskxi[idx_rmax:idx_rmin+1] = 1
    mask = np.hstack([maskp, maskxi])
    Nk = np.sum(mask)
    
    m1, m2 = np.mgrid[0:cov.shape[0],0:cov.shape[0]]
    mask_2d = mask[m1] * mask[m2]
    
    print 'N ', Nk, ' kmin', idx_kmin, ' kmax', idx_kmax, ' rmin', idx_rmin, ' rmax', idx_rmax
    
    #covinv = np.linalg.inv(cov[mask_2d].reshape(Nk, Nk))
    covp = cov[idx_kmin:idx_kmax+1, idx_kmin:idx_kmax+1]
    covxi = cov[RSDPower.kcenter_y.size + idx_rmax:RSDPower.kcenter_y.size+idx_rmin+1,\
                RSDPower.kcenter_y.size + idx_rmax:RSDPower.kcenter_y.size+idx_rmin+1]
    covpxi = cov[idx_kmin:idx_kmax+1,\
                RSDPower.kcenter_y.size + idx_rmax:RSDPower.kcenter_y.size+idx_rmin+1]
    covinv = invert_covtot(covp ,covxi, covpxi)
    
    #print covinv.diagonal
    #print np.sum( covinv.diagonal() < 0)
    
    stop
    fig, ax = plt.subplots()
    
    bestfit_b = []
    chi2result = np.zeros(blist.size)
    for i in range(mockdatavs.shape[1]):
    #for i in range(100):
        print '{}/{} \r'.format(i+1, mockdatavs.shape[1] ),
        for j in range(blist.size) : 
            dv = datavs[j][mask]
            mv = mockdatavs[:,i][mask]
            chi2result[j] = best_chisqr( covinv = covinv, datav=dv, mockv = mv)
        maxind = np.argmin(chi2result)
        bestfit_b.append(blist[maxind])
        print chi2result[maxind]
        ax.plot(blist, chi2result, color = 'grey', alpha = 0.2)
        
    print ' sigma_b :', np.std(bestfit_b)
    return bestfit_b





def getting_sigma_bs_diff( RSDPower, covp = None, covxi = None, 
                          datavsp = None, datavsxi = None, 
                          mockdatavsp = None, mockdatavsxi = None, p=False, 
                          kmin=None, kmax = None,
                          rmin = None, rmax = None):
    
    blist = np.linspace(1.5, 2.5, 100)
    
    if kmin == None : 
        idx_kmin = 0
        idx_kmax = RSDPower.kmax_y.size
    else : 
        idx_kmin = get_closest_index_in_data( kmin, RSDPower.kmin_y )   
        idx_kmax = get_closest_index_in_data( kmax, RSDPower.kmax_y )
        
    if rmin == None : 
        idx_rmin = RSDPower.rmax.size
        idx_rmax = 0
    else : 
        idx_rmin = get_closest_index_in_data( rmin, RSDPower.rmin )   
        idx_rmax = get_closest_index_in_data( rmax, RSDPower.rmax )

        
    maskp = np.zeros(RSDPower.kcenter_y.size, dtype=bool)
    maskp[idx_kmin:idx_kmax+1] = 1
    maskxi = np.zeros(RSDPower.rcenter.size, dtype=bool)   
    maskxi[idx_rmax:idx_rmin+1] = 1
    Nk, Nr = np.sum(maskp), np.sum(maskxi) 
        
    m1, m2 = np.mgrid[0:covp.shape[0],0:covp.shape[0]]
    maskp_2d = maskp[m1] * maskp[m1].T
    m1, m2 = np.mgrid[0:covxi.shape[0],0:covxi.shape[0]]
    maskxi_2d = maskxi[m1] * maskxi[m1].T    
    
    print 'Nk ', Nk, ' kmin', idx_kmin, ' kmax', idx_kmax
    print 'Nr ', Nr, ' rmin', idx_rmin, ' rmax', idx_rmax
    
    
    covinv_p = np.linalg.inv(covp[maskp_2d].reshape(Nk, Nk))
    covinv_xi = np.linalg.inv(covxi[maskxi_2d].reshape(Nr, Nr))
    
    bestfit_b = []
    chi2result = np.zeros(blist.size)
    chi2result1 = np.zeros(blist.size)
    chi2result2 = np.zeros(blist.size)
    
    fig, ax = plt.subplots()
    for i in range(mockdatavsp.shape[1]):
    #for i in range(10):
        print '{}/{} \r'.format(i+1, mockdatavsp.shape[1] ),
        for j in range(blist.size) : 
            
            dv = datavsp[j][maskp]
            mv = mockdatavsp[:,i][maskp]
            dvxi = datavsxi[j][maskxi]
            mvxi = mockdatavsxi[:,i][maskxi]
            chi1 = best_chisqr( covinv = covinv_p, datav=dv, mockv = mv, p=True)
            chi2 = best_chisqr( covinv = covinv_xi, datav=dvxi, mockv = mvxi)
            chi2result[j] = chi1+chi2
            chi2result1[j] = chi1
            chi2result2[j] = chi2
        maxind = np.argmin(chi2result)
        bestfit_b.append(blist[maxind])
        print chi2result[maxind]
        ax.plot(blist, chi2result, color = 'grey', alpha = 0.2)
        #ax.plot(blist, chi2result, color = 'blue', alpha = 0.2)
        #ax.plot(blist, chi2result, color = 'red', alpha = 0.2)
    
        #stop
    
    print ' sigma_b :', np.std(bestfit_b)
    return bestfit_b
```

```python
# two k bin
    
def check_consistency( F, kind, kN, rind, rN):
    
    krange =  'k = {:10} ~ {:10}'.format(cosmo.kcenter_y[kind], cosmo.kcenter_y[kind+kN])
    rrange = 'r = {:10} ~ {:10}'.format(cosmo.rcenter[rind+rN], cosmo.rcenter[rind])

    if F is None : 
        print krange
        print rrange

    dpdb = dpdb_model[kind:kind+kN]
    dxidb = dxidb_model[rind:rind+rN]
    dcdb = np.hstack((dpdb, dxidb))

    cov_p = covp_model[kind:kind+kN, kind:kind+kN]
    cov_xi = covxi_model[rind:rind+rN, rind:rind+rN]
    cov_pxi = covpxi_model[kind:kind+kN, rind:rind+rN]
    cov_tot = np.vstack(( np.hstack((cov_p, cov_pxi)), np.hstack((cov_pxi.T, cov_xi)) ))


    from numpy.linalg import inv

    F_pb = np.dot( np.dot( dpdb, inv(cov_p) ), dpdb.T )
    F_xb = np.dot( np.dot( dxidb, inv(cov_xi) ), dxidb.T )
    F_cb = np.dot( np.dot( dcdb, inv(cov_tot) ), dcdb.T )
    F_db = F_pb + F_xb

    if np.sum( inv(cov_tot).diagonal() < 0) > 0 : print 'Warning : negative Fisher diagonals'
        
    sigma_pb = np.sqrt(1./F_pb)
    sigma_xb = np.sqrt(1./F_xb)
    sigma_cb = np.sqrt(1./F_cb)
    sigma_db = np.sqrt(1./F_db)

    if sigma_cb < sigma_db : 
        if F is None : print "Failed : Something must be wrong!!!"
        mask = 1
    else : 
        if F is None : print 'Success'
        mask = 0
    
    if F is None : 
        print '--------------------------------------'
        print 'sigma_p    = ',sigma_pb
        print 'sigma_xi   = ',sigma_xb
        print 'sigma_com  = ',sigma_cb
        print 'sigma_diff = ',sigma_db
        print ''
    else :
        F.write('{} {} {}\n'.format(cosmo.rcenter[rind+rN], cosmo.rcenter[rind], mask))
        F.flush()  

```
